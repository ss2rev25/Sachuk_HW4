```python
museums = ['avila_adobe', 'firehouse_museum', 'chinese_american_museum', 'america_tropical_interpretive_center']
visitors_2015 = df[df.date.dt.year == 2015]
month = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
print("Month with bigest total visitors in 2015: ", month[visitors_2015.drop(columns="date").idxmax()[0]-12])
print("Month with lowest total visitors in 2015: ", month[visitors_2015.drop(columns="date").idxmin()[0]-12])
```

13. Побудувати графік ліній для кожного музею, показуючи їхню відвідуваність протягом 2017 року.

```python
df_2017 = df[df.date.dt.year == 2017]
plt.figure(figsize=(16, 6))
for i in museums:
    plt.plot(df_2017.date, df_2017[i], label=i)

plt.title("Attendance in 2017")
plt.xlabel("Months")
plt.ylabel("Number of visitors")
plt.legend()
plt.show()
```

14. Побудувати графіки розсіювання для кожного музею за 2018 рік

```python
df_2018 = df[df.date.dt.year == 2018]
plt.figure(figsize=(16, 6))
for i in museums:
    plt.scatter(df_2018.date, df_2018[i], label=i, alpha=0.5)

plt.title("Attendance in 2018")
plt.xlabel("Months")
plt.ylabel("Number of visitors")
plt.legend()
plt.show()
```

15. Гістограма відвідуваності за місяцями кожного музею. Вивести 4 графіки: за 2014, 2015, 2016, 2017 роки. Графіки розташовані два зверху, два знизу, колір першого графіка - синій, другого зелений, третього жовтий, четвертого сірий.

```python
years = [2014, 2015, 2016, 2017]
colors = ['red', 'green', 'yellow', 'blue']
fig, axes = plt.subplots(2, 2, figsize=(16, 10))

for i, year in enumerate(years):
    df_year = df[df.date.dt.year == year]
    monthly_visitors = df_year.groupby(df_year.date.dt.month)[museums].sum()

    for o, museum in enumerate(museums):
        ax = axes[i // 2, i % 2]
        ax.hist(monthly_visitors[museum], color=colors[o], label=museum, alpha=0.5)

        ax.set_title(f"{year}")
        ax.legend()

plt.tight_layout()
plt.subplots_adjust()
plt.show()
```
